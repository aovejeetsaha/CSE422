# 🍄 Mushroom Dataset EDA & Cleaning
This notebook performs Exploratory Data Analysis (EDA) on the mushroom dataset. We first analyze the raw data to spot issues, then clean it and re-run EDA.

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")
```

## 1. Load Dataset

```

# Load CSV with correct separator
file_path = "mushroom_dataset.csv"
df = pd.read_csv(file_path, sep=";")
df.columns = [c.strip() for c in df.columns]

print("Dataset shape:", df.shape)
df.head()
```

## 2. EDA on Raw Data

```

# Dataset info
df.info()
```

```

# Missing values
df.isna().sum()
```

```

# Visualize missing values
plt.figure(figsize=(12,6))
sns.heatmap(df.isna(), cbar=False)
plt.title("Missing Values Heatmap (Raw Data)")
plt.show()
```

```

# Class distribution
plt.figure(figsize=(6,4))
sns.countplot(x="class", data=df, palette="Set2")
plt.title("Class Distribution (Raw Data)")
plt.show()
```

```

# Example categorical feature distribution
plt.figure(figsize=(8,4))
sns.countplot(x="cap-shape", data=df, order=df["cap-shape"].value_counts().index)
plt.title("Cap Shape Distribution (Raw Data)")
plt.show()
```

```

# Summary statistics for numeric features
df.describe()
```

## 3. Data Cleaning

```

# Drop columns with too many missing values
df_clean = df.drop(columns=["veil-type", "veil-color", "spore-print-color"])

# Handle missing values (simple imputation with mode for categorical, mean for numeric)
for col in df_clean.columns:
    if df_clean[col].dtype == "object":
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    else:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mean())

print("Dataset shape after cleaning:", df_clean.shape)
df_clean.isna().sum().sum()
```

## 4. EDA on Cleaned Data

```

# Check missing values again
df_clean.isna().sum()
```

```

# Class distribution after cleaning
plt.figure(figsize=(6,4))
sns.countplot(x="class", data=df_clean, palette="Set1")
plt.title("Class Distribution (Cleaned Data)")
plt.show()
```

```

# Example feature distribution after cleaning
plt.figure(figsize=(8,4))
sns.countplot(x="cap-color", data=df_clean, order=df_clean["cap-color"].value_counts().index)
plt.title("Cap Color Distribution (Cleaned Data)")
plt.show()
```

```

# Correlation heatmap for numeric features
plt.figure(figsize=(8,6))
sns.heatmap(df_clean.select_dtypes(include=[np.number]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```
